# 7. Which brand had the largest absolute € difference in average transaction value between domestic and international?

```python
A = df[df["seller_country"] == df["buyer_country"]].groupby("brand").sum("euro")[["euro"]]
A    
```

```python
B = df[df["seller_country"] != df["buyer_country"]].groupby("brand").sum("euro")[["euro"]]
B
```

```python
A - B
```

```python
(A-B).sort_values("euro", ascending=False).head(1)
```
