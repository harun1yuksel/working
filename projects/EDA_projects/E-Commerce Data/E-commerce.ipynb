# Count the total number of unique customers in January 2011 and how many of them came back every month over the entire year in 2011 

```python
df.T
```

```python
df[(df["Order_Date"].dt.year == 2011) & (df["Order_Date"].dt.month == 1)]
```

```python
df[(df["Order_Date"].dt.year == 2011) & (df["Order_Date"].dt.month == 1)]["Cust_id"].nunique()
```

```python
for i in list(range(1, 13)):
    customers_2011jan = df[(df["Order_Date"].dt.year == 2011) & (df["Order_Date"].dt.month == 1)]["Cust_id"]
    if i == 1:
        print("count of customers in Jan 2011 is", customers_2011jan.nunique())        
        print("--------------")
    else: 
       a = df[(df["Order_Date"].dt.year == 2011) & (df["Order_Date"].dt.month == i) & (df.Cust_id.isin(customers_2011jan))]["Cust_id"].nunique()
       print(round(a / customers_2011jan.nunique() * 100, 1),  "percent of them came back in",  datetime.date(1900, i , 1).strftime('%B'))
```
